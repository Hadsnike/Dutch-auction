```python
from tkinter import *
import random
import time
from tkinter import messagebox
import random

root = Tk()
root.title("Голландский аукцион")
w = root.winfo_screenwidth() 
h = root.winfo_screenheight() 
colors = ["red", "green", "blue", "yellow", "orange", "purple"]
count = 10

start_price = 10000
start_tovar = 100
bot1_money, bot2_money, bot3_money = random.random() * 10000, random.random() * 10000, random.random() * 10000

def price_change():
    global start_price, bot1_money, bot2_money, bot3_money, rand1, rand2, rand3
    start_price -= 1000
    rand1 = random.randint(1,3)
    rand2 = random.randint(1,3)
    rand3 = random.randint(1,3)
    while len(set([rand1, rand2, rand3]))<3:
        rand1 = random.randint(1,3)
        rand2 = random.randint(1,3)
        rand3 = random.randint(1,3)
    if start_price <= 10000:
        l1.config(text= "Новая цена: {}".format (start_price))
    if start_price <= bot1_money and rand1 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 1 готов совершить покупку")
        l2.config(text= "Участник под номером 1 готов купить")
    if start_price <= bot2_money and rand2 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 2 готов совершить покупку")
        l3.config(text= "Участник под номером 2 готов купить")
    if start_price <= bot3_money and rand3 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 2 готов совершить покупку")
        l4.config(text= "Участник под номером 3 готов купить")
        

def prodaja():
    global start_price, bot1_money, bot2_money, bot3_money, rand1, rand2, rand3, count
    if start_price <= bot1_money and rand1 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 1")
    if start_price <= bot2_money and rand2 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 2")
    if start_price <= bot3_money and rand3 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 3")
    count-=1
    if count==0: b1['command'] = 'disabled'

    
l1= Label(text = "Изначальная цена: {}".format(start_price), anchor = W, width = 35)
l1.grid(row= 1, column= 0, columnspan= 3)
l2= Label(text = "Участник под номером 1 не готов купить", anchor = W, width = 35)
l2.grid(row= 1, column= 3, columnspan= 3)
l3= Label(text = "Участник под номером 2  не готов купить", anchor = W, width = 35)
l3.grid(row= 2, column= 3, columnspan= 3)
l4= Label(text = "Участник под номером 3  не готов купить", anchor = W, width = 35)
l4.grid(row= 3, column= 3, columnspan= 3)


b1= Button(text = "Понизить цену", width= 12, command = price_change)
b1.grid(row= 3, column= 0)
b2= Button(text = "Продать", width= 12, bg = 'red', command = prodaja)
b2.grid(row= 3, column= 1)
root.mainloop()
```
