```python
from tkinter import *
import random
import time
from tkinter import messagebox, Label
import random

root = Tk()
root.title("Голландский аукцион")
w = root.winfo_screenwidth() 
h = root.winfo_screenheight() 
colors = ["red", "green", "blue", "yellow", "orange", "purple"]

tovar = 10
start_price = 10000 / (tovar / 10)
price = 10000 / (tovar / 10)
bot1_money, bot2_money, bot3_money = int(random.random() * 10000), int(random.random() * 10000), int(random.random() * 10000)
while bot1_money <= 1000 or bot2_money <= 1000 or bot3_money <= 1000:
    bot1_money, bot2_money, bot3_money = int(random.random() * 10000), int(random.random() * 10000), int(random.random() * 10000)


def reshimost():
    global rand
    rand = [random.randint(1,3) for x in range(3)]
    while len(set(rand)) < 3:
        rand = [random.randint(1,3) for x in range(3)]

money = {1: bot1_money, 2: bot2_money, 3: bot3_money}
bot = {1: 0, 2: 0, 3: 0}
  
 

l = [
  Label(root, text=""), 
  Label(root, text=""), 
  Label(root, text=""),
  Label(root, text="")
]

def отрисовка():
    for i in range(1,4):
        if money.get(i) > 1000:
            if price > money.get(i):
                l[i].config(text= "Участник под номером {} не может купить      Капитал: {}".format(i, money.get(i)), font='Times 18')
            else: 
                if rand[i-1] == 1:
                    messagebox.showwarning("Уведомление о торгах", "Участник под номером {} готов совершить покупку".format(i))
                    l[i].config(text= "Участник под номером {} готов купить".format(i), font='Times 18')
                    b2['state']='normal'
                else: 
                    l[i].config(text= "Участник под номером {} может, но не хочет  Капитал: {}".format(i, money.get(i)),  font='Times 18')
def положняк():
    global rand
    l1.config(text= "Новая цена: {}".format(price), font='Times 18')
    отрисовка()
            
reshimost()
def price_change():
    global price, bot1_money, bot2_money, bot3_money, rand, start_price
    price -= int(price * 0.05)
    reshimost()
    положняк()
        

def prodaja():
    global price, money, rand, tovar, start_price, bot_count
    for i in range(1,4):
        if price <= money.get(i) and rand[i-1] == 1:
            #essagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером {}".format(i))
            if money.get(i)<=1000:
                messagebox.showerror("Уведомление о выходе из торгов", "Участник под номером {} выходит из торгов".format(i))            
            money[i] -= price
            bot[i] += price
            b2['state']='disabled'
        if money.get(1)<=1000 and money.get(2)<=1000 and money.get(3)<=1000:
            b1['state']='disabled'
    tovar -= 1
    l5.config(text= "Текущее количество товара: {}".format (tovar), font='Times 18')
    if tovar == 0: 
        b1['command'] = 'disabled'
        b2['command'] = 'disabled'
    else:
        l5.config(text= "Текущее количество товара: {}".format (tovar), font='Times 18')
        start_price = int(start_price / (tovar / 10))
        price = start_price
        положняк()
        l7.config(text = "Участник под номером 1 потратил {} долларов".format(bot[1]), font='Times 18')
        l8.config(text = "Участник под номером 2 потратил {} долларов".format(bot[2]), font='Times 18')
        l9.config(text = "Участник под номером 3 потратил {} долларов".format(bot[3]), font='Times 18')
        for i in range(1,4):
            if money.get(i) <= 1000:
                l[i].config(text= "Участник под номером {} покинул торги".format(i), font='Times 18')
        
    

l1= Label(text = "Изначальная цена: {}".format(start_price), anchor = W, width = 35, font='Times 18')
l1.grid(row= 1, column= 0, columnspan= 3)

for i, label in enumerate(l):
    label.grid(row=i, column= 3, columnspan= 3)
отрисовка()

l5= Label(text = "Текущее количество товара: {}".format(tovar), anchor = W, width = 25, font='Times 18')
l5.grid(row= 2, column= 0, columnspan= 1)
l7= Label(text = "Участник под номером 1 потратил {} долларов".format(bot[1]), anchor = W, width = 50, font = 'Times 18')
l7.grid(row= 5, column= 4, columnspan= 1)
l8= Label(text = "Участник под номером 2 потратил {} долларов".format(bot[2]), anchor = W, width = 50, font = 'Times 18')
l8.grid(row= 6, column= 4, columnspan= 1)
l9= Label(text = "Участник под номером 3 потратил {} долларов".format(bot[3]), anchor = W, width = 50, font = 'Times 18')
l9.grid(row= 7, column= 4, columnspan= 1)


b1= Button(text = "Понизить цену", width= 12, command = price_change, font='Times 18')
b1.grid(row= 7, column= 0)
b2= Button(text = "Продать", width= 12, bg = 'red', command = prodaja, font='Times 18')
b2.grid(row= 7, column= 1)

if price==start_price:
    b2['state']='disabled'

root.mainloop()
```
