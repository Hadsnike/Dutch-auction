```python
from tkinter import *
import random
import time
from tkinter import messagebox
import random

root = Tk()
root.title("Голландский аукцион")
w = root.winfo_screenwidth() 
h = root.winfo_screenheight() 
colors = ["red", "green", "blue", "yellow", "orange", "purple"]

tovar = 10
start_price = 10000 / (tovar / 10)
price = 10000 / (tovar / 10)
viruchka = 0
bot1_money, bot2_money, bot3_money = int(random.random() * 10000), int(random.random() * 10000), int(random.random() * 10000)
while bot1_money <= 1000 or bot2_money <= 1000 or bot3_money <= 1000:
    bot1_money, bot2_money, bot3_money = int(random.random() * 10000), int(random.random() * 10000), int(random.random() * 10000)
bot1_s = 0
bot2_s = 0
bot3_s = 0

def price_change():
    global price, bot1_money, bot2_money, bot3_money, rand1, rand2, rand3, start_price, viruchka
    price -= int(price * 0.05)
    rand1 = random.randint(1,3)
    rand2 = random.randint(1,3)
    rand3 = random.randint(1,3)
    while len(set([rand1, rand2, rand3])) < 3:
        rand1 = random.randint(1,3)
        rand2 = random.randint(1,3)
        rand3 = random.randint(1,3)   
    l1.config(text= "Новая цена: {}".format(price), font='Times 18')
    if bot1_money > 1000:
        l2.config(text= "Участник под номером 1 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot1_money, rand1), font='Times 18')
    if bot2_money > 1000:
        l3.config(text= "Участник под номером 2 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot2_money, rand2), font='Times 18')
    if bot3_money > 1000:
        l4.config(text= "Участник под номером 3 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot3_money, rand3), font='Times 18')
    if price <= bot1_money and rand1 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 1 готов совершить покупку")
        l2.config(text= "Участник под номером 1 готов купить", font='Times 18')
    elif price <= bot2_money and rand2 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 2 готов совершить покупку")
        l3.config(text= "Участник под номером 2 готов купить", font='Times 18')
    elif price <= bot3_money and rand3 == 1:
        messagebox.showwarning("Уведомление о торгах", "Участник под номером 3 готов совершить покупку")
        l4.config(text= "Участник под номером 3 готов купить", font='Times 18')
        

def prodaja():
    global price, bot1_money, bot2_money, bot3_money, rand1, rand2, rand3, tovar, start_price, viruchka, bot1_s, bot2_s, bot3_s
    if price <= bot1_money and rand1 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 1")
        bot1_money -= price
        bot1_s += price
    if price <= bot2_money and rand2 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 2")
        bot2_money -= price
        bot2_s += price
    if price <= bot3_money and rand3 == 1:
        messagebox.showinfo("Уведомление о продаже", "Товар успешно продан участнику под номером 3")
        bot3_money -= price
        bot3_s += price
    tovar -= 1
    l5.config(text= "Текущее количество товара: {}".format (tovar), font='Times 18')
    if tovar == 0: 
        b1['command'] = 'disabled'
        b2['command'] = 'disabled'
    else:
        l5.config(text= "Текущее количество товара: {}".format (tovar), font='Times 18')
        viruchka += price
        start_price = int(start_price / (tovar / 10))
        price = start_price
        l1.config(text= "Новая цена: {}".format (start_price), font='Times 18')
        if bot1_money > 1000:
            l2.config(text= "Участник под номером 1 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot1_money, rand1), font='Times 18')
        if bot2_money > 1000:
            l3.config(text= "Участник под номером 2 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot2_money, rand2), font='Times 18')
        if bot3_money > 1000:
            l4.config(text= "Участник под номером 3 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot3_money, rand3), font='Times 18')
        l6.config(text= "Текущая выручка: {}".format (viruchka), font='Times 18')
        l7.config(text = "Участник под номером 1 потратил {} долларов".format(bot1_s), font='Times 18')
        l8.config(text = "Участник под номером 1 потратил {} долларов".format(bot2_s), font='Times 18')
        l9.config(text = "Участник под номером 1 потратил {} долларов".format(bot3_s), font='Times 18')
        if bot1_money <= 1000:
            messagebox.showerror("Уведомление о выходе из торгов", "Участник под номером 1 выходит из торгов")
            l2.config(text= "Участник под номером 1 покинул торги", font='Times 18')
        if bot2_money <= 1000:
            messagebox.showerror("Уведомление о выходе из торгов", "Участник под номером 2 выходит из торгов")
            l3.config(text= "Участник под номером 2 покинул торги", font='Times 18')
        if bot3_money <= 1000:
            messagebox.showerror("Уведомление о выходе из торгов", "Участник под номером 3 выходит из торгов")
            l4.config(text= "Участник под номером 3 покинул торги", font='Times 18')
        
    
    
l1= Label(text = "Изначальная цена: {}".format(start_price), anchor = W, width = 35, font='Times 18')
l1.grid(row= 1, column= 0, columnspan= 3)
l2= Label(text = "Участник под номером 1 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot1_money, rand1), anchor = W, width = 70, font='Times 18')
l2.grid(row= 1, column= 3, columnspan= 3)
l3= Label(text = "Участник под номером 2 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot2_money, rand2), anchor = W, width = 70, font='Times 18')
l3.grid(row= 2, column= 3, columnspan= 3)
l4= Label(text = "Участник под номером 3 не готов купить    Капитал: {}  Индекс решимости: {}".format(bot3_money, rand3), anchor = W, width = 70, font='Times 18')
l4.grid(row= 3, column= 3, columnspan= 3)
l5= Label(text = "Текущее количество товара: {}".format(tovar), anchor = W, width = 25, font='Times 18')
l5.grid(row= 2, column= 0, columnspan= 1)
l6= Label(text = "Текущая выручка: {}".format(viruchka), anchor = W, width = 25, font='Times 18')
l6.grid(row= 3, column= 0, columnspan= 1)
l7= Label(text = "Участник под номером 1 потратил {} долларов".format(bot1_s), anchor = W, width = 50, font = 'Times 18')
l7.grid(row= 5, column= 4, columnspan= 1)
l8= Label(text = "Участник под номером 2 потратил {} долларов".format(bot2_s), anchor = W, width = 50, font = 'Times 18')
l8.grid(row= 6, column= 4, columnspan= 1)
l9= Label(text = "Участник под номером 3 потратил {} долларов".format(bot3_s), anchor = W, width = 50, font = 'Times 18')
l9.grid(row= 7, column= 4, columnspan= 1)


b1= Button(text = "Понизить цену", width= 12, command = price_change, font='Times 18')
b1.grid(row= 7, column= 0)
b2= Button(text = "Продать", width= 12, bg = 'red', command = prodaja, font='Times 18')
b2.grid(row= 7, column= 1)
root.mainloop()
```
